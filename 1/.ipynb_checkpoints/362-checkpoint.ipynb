```python
def fit_curve(x,y):
    print ('Fitting curve ... ')
    print ('x = ',x)
    print ('y = ',y)
    x = np.array(x)
    y = np.array(y)
    k, b = np.polyfit(np.log(x), y, 1)
    print ('Curve found: y = {}*np.log(x) + {}'.format(k,b))
    r = []
    for n in x:
        r.append(k * math.log(n, math.e) + b)
    error = max([(a0-a)/a for (a0,a) in zip(r, y)])
    print ('Error of curve = ', error)
```
